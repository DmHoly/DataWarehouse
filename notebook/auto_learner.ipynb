```python
from pycaret.datasets import get_data
```

```python
df_dataset = get_data('index')
df_dataset['Dataset'].unique()
```

```python
dataset_for_training = get_data('diamond')
```

```python
# Import necessary libraries
from pycaret.regression import setup
setup(data = dataset_for_training, target = 'Price')
```

```python
# compare model 
from pycaret.regression import compare_models
bestmodel = compare_models()
```

```python
#print the result of the model
bestmodel 

#plot the model
from pycaret.regression import plot_model

plot_model(bestmodel, plot = 'residuals')
```

```python
plot_model(bestmodel, plot = 'feature')
```

```python
plot_model(bestmodel, plot = 'error')
```
